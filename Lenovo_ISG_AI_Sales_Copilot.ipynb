```python
# To demonstate all kinds of different tokens, let's us print out tokens from the longest to the shortest.

tokens_with_length = [] #Store the token ID, token, and its length.


for token_id in range(tokenizer.vocab_size): #Enumerate token id
    token = tokenizer.decode(token_id) #Convert token_id to token
    tokens_with_length.append((token_id, token, len(token))) #len(token) is the length of the token

# Sort the tokens from longest to the shorest
tokens_with_length.sort(key=lambda x: x[2], reverse=True) #If we change reverse=True to reverse=False, then we sort from shortest to longest.

# Print out the first 100 longest tokens
k = 100
for t in range(k):
    token_id, token_str, token_length = tokens_with_length[t]
    print("Token ID ", token_id, " (length: ", token_length, ")", tokenizer.decode(token_id))
```

```python
prompt = "1+1="
print("Input prompt is:", prompt)

input_ids = tokenizer.encode(prompt, return_tensors="pt") # return_tensors="pt" means PyTorch tensor format
print("This is what model can read：",input_ids)

# model takes input_ids as inputs and generates outputs，
outputs = model(input_ids)

# outputs contain lots of information and you may ignore the following two lines of code if you are not sure
# outputs.logits contains the score for each word to be the next token at each position of the prompt
# outputs.logits shape: (batch_size, sequence_length, vocab_size)
last_logits = outputs.logits[:, -1, :] #get the score for each word to the next token at the end of the prompt
probabilities = torch.softmax(last_logits, dim=-1) #softmax converts scores into probabilities

# Print out the top_k mostly likely tokens
top_k = 10
top_p, top_indices = torch.topk(probabilities, top_k)
print(f"The top {top_k} most likely token:")
for i in range(top_k):
    token_id = top_indices[0][i].item() # Take the i-th most likely token ID
    probability = top_p[0][i].item() # Get the corresponding probability
    token_str = tokenizer.decode(token_id) # Decode token ID to token
    print(f"Token ID: {token_id}, Token: '{token_str}', Probability: {probability:.4f}")
```

```python
prompt = "Who are you?"
length = 16 #Generate a sequence of 16 tokens

for t in range(length):
  print("Input token is:", prompt)
  input_ids = tokenizer.encode(prompt,return_tensors="pt")

  # Use the model to generate the next token
  outputs = model(input_ids)
  last_logits = outputs.logits[:, -1, :]
  probabilities = torch.softmax(last_logits, dim=-1)
  top_p, top_indices = torch.topk(probabilities, 1)
  token_id = top_indices[0][0].item() # Take the mostly likely token ID
  token_str = tokenizer.decode(token_id) #token_str is the token
  print("Next token is:", token_str)

  prompt = prompt + token_str #Add the new token to the end of prompt, and use it as the new prompt for next iteration.
```

```python
prompt = "Who are you?"
length = 16 #Generate a sequence of 16 tokens

for t in range(length):
  print("Input token is:", prompt)
  input_ids = tokenizer.encode(prompt,return_tensors="pt")

  # Use the model to generate the next token
  outputs = model(input_ids)
  last_logits = outputs.logits[:, -1, :]
  probabilities = torch.softmax(last_logits, dim=-1)

  #top_p, top_indices = torch.topk(probabilities, 1)
  #token_id = top_indices[0][0].item() # Take the mostly likely token ID
  token_id = torch.multinomial(probabilities, num_samples=1).squeeze() #use the probabilities to randomly sample a token

  token_str = tokenizer.decode(token_id)
  print("Next token is：\n", token_str)

  prompt = prompt + token_str #Add the new token to the end of prompt, and use it as the new prompt for next iteration.
```

```python
prompt = "Who are you?"
length = 16 #Generate a sequence of 16 tokens

for t in range(length):
  print("Input token is:", prompt)
  input_ids = tokenizer.encode(prompt,return_tensors="pt")

  # Use the model to generate the next token
  outputs = model(input_ids)
  last_logits = outputs.logits[:, -1, :]
  probabilities = torch.softmax(last_logits, dim=-1)

  #top_p, top_indices = torch.topk(probabilities, 1)
  #token_id = top_indices[0][0].item() # Take the mostly likely token ID
  #token_id = torch.multinomial(probabilities, num_samples=1).squeeze() #use the probabilities to randomly sample a token

  top_p, top_indices = torch.topk(probabilities, top_k) #First find the top-k most likely tokens
  sampled_index = torch.multinomial(top_p.squeeze(0), num_samples=1).item() #randomly sample one from the top-k according to the probability
  token_id = top_indices[0][sampled_index].item() #Find the sampled token ID

  token_str = tokenizer.decode(token_id) #Decode back to token
  print("Next token is:", token_str)
  prompt = prompt + token_str #Add the new token to the end of prompt, and use it as the new prompt for next iteration.
```

```python
import torch

# Store chat history
messages = []

# System Prompt
messages.append({
    "role": "system",
    "content": "Your name is Llama. Please respond to questions concisely."
})

# Multi-turn conversation
while True:

    # User input
    user_prompt = input("😊 You say: ")

    if user_prompt.lower() == "exit":
        print("👋 Chat ended.")
        break

    # Add user message
    messages.append({
        "role": "user",
        "content": user_prompt
    })

    # Convert chat history to model input
    inputs = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        return_tensors="pt",
        return_dict=True
    )

    # Move tensors to GPU if model is on GPU
    inputs = {
        k: v.to(model.device)
        for k, v in inputs.items()
    }

    # Generate response
    outputs = model.generate(
        input_ids=inputs["input_ids"],
        attention_mask=inputs["attention_mask"],
        max_new_tokens=200,
        do_sample=True,
        top_k=3,
        pad_token_id=tokenizer.eos_token_id
    )

    # Extract only newly generated tokens
    response_ids = outputs[0][inputs["input_ids"].shape[-1]:]

    response = tokenizer.decode(
        response_ids,
        skip_special_tokens=True
    ).strip()

    # Print response
    print("🤖 Assistant says:", response)

    # Save response into history
    messages.append({
        "role": "assistant",
        "content": response
    })
```

```python
system_prompt = """
You are Lenovo ISG AI Sales Copilot.

Your mission is to help Lenovo sales teams improve sales efficiency using AI.

For every customer request, complete the following tasks:

1. Customer Summary
- Summarize the customer's business background.

2. Business Challenges
- Identify possible business pain points.

3. Recommended Lenovo ISG Solution
Recommend a high-level Lenovo AI solution architecture.
Focus on business capabilities instead of specific products.
Your recommendation should include:
- AI Infrastructure Strategy
- GPU Resource Planning
- High-performance Storage Architecture
- Private AI Strategy
- Hybrid AI Deployment Strategy
- Security & Compliance
- AI Lifecycle Management
- Expected Business Value
Do NOT recommend any specific Lenovo server, storage, GPU, or hardware model names.
If you are uncertain about a product, describe the capability instead.
Focus on solution architecture and business value rather than product catalog.

4. Sales Strategy
- Suggest how the sales representative should approach the customer.

5. Next Follow-up Actions
- Suggest practical next steps.
Requirements:
- Be concise.
- Be professional.
- Use business language.
- Think step by step.

6. AI Opportunity Assessment
At the end of your analysis, provide:
- Opportunity Score (0-100)
- Priority (High / Medium / Low)
- Three business reasons
- Recommended next sales action

Important Rules:
1. Only recommend Lenovo ISG solutions.
2. Never recommend competitors such as Dell, HPE, Cisco or Supermicro.
3. If you are not sure about a specific Lenovo product, do not make up a product name.
4. Instead, recommend a Lenovo AI infrastructure solution.
"""

print(system_prompt)
```

```python
def analyze_customer(industry,
                     company,
                     country,
                     budget,
                     need,
                     challenge):
    """
    Analyze one customer using Lenovo ISG AI Sales Copilot.
    """

    customer_info = f"""
Customer Information

Industry: {industry}

Company: {company}

Country: {country}

Budget: {budget}

Need:
{need}

Current Challenge:
{challenge}
"""

    prompt = system_prompt + "\n\n" + customer_info

    inputs = tokenizer(prompt, return_tensors="pt")

    outputs = model.generate(
        **inputs,
        max_new_tokens=900,
        temperature=0.7,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )

    response = tokenizer.decode(
        outputs[0],
        skip_special_tokens=True
    )

    return response
```

```python
def load_customer(industry):

    customer = customer_database[industry]

    return (
        customer["company"],
        customer["country"],
        customer["budget"],
        customer["need"],
        customer["challenge"]
    )
```
